# 데이터 확인과 예측 실행

```python
# 데이터셋 로드 및 확인
import pandas as pd

!git clone https://github.com/taehojo/data.git

df = pd.read_csv('./data/sonar3.csv', header=None)

df.head()
```

```python
# 클래스 확인 (일반 암석 & 광석)
df[60].value_counts()

# 광석이 111개, 일반 암석이 97개로 총 208개의 샘플 존재
```

```python
# 데이터 분할
X = df.iloc[:, 0:60]
y = df.iloc[:, 60]
```

```python
# 예측 실행

# 모델 생성
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()

model.add(Dense(24, input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# 모델 컴파일
model.compile(loss='binary_crossentropy', optimizer='adam',
              metrics=['accuracy'])

# 모델 실행
history = model.fit(X, y, epochs=200, batch_size=10)

# 결과를 보니 정확도가 100%임 --> 과대적합된 것이 아닐까?
```

# 과적합

* **과적합(overfitting)**이란 모델이 훈련 세트에서는
<br> 일정 수준 이상의 예측 정확도를 보이지만,
<br> 새로운 데이터에 적용하면 잘 맞지 않는 것을 의미한다.

# 훈련 세트와 테스트 세트

* 과적합을 방지하려면 어떻게 해야 할까?

* 데이터셋을 훈련(학습) 세트와 테스트 세트로 분할 후
<br> 학습과 동싱 테스트를 병행하여 진행하면 된다.

* 예를 들어 데이터셋이 총 100개의 샘플로 구성되어 있다면,
<br> 70%는 훈련 세트로 30%는 테스트 세트로 나눈다.

* 훈련 세트는 훈련이 진행됨에 따라(epoch 증가) 에러가 작아지지만, 테스트 세트에서는 과적합이 발생한다.

```python
# 데이터셋을 훈련 세트와 테스트 세트로 분할
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True)
```

```python
# 훈련 세트 모델 재생성
model = Sequential()
model.add(Dense(24,  input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# 모델 컴파일
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# 모델 실행
history=model.fit(X_train, y_train, epochs=200, batch_size=10)
```

```python
# 앞서 만든 모델 구조를 유지하고 테스트 실행
score=model.evaluate(X_test, y_test)

print('Test accuracy:', score[1])

# evaluate() 함수는 만들어진 모델을 테스트셋에 적응시킴
# 훈련 세트의 정확도는 100%인 반면, 테스트 세트의 정확도는 84%임 --> 과대적합 (테스트 세트의 크기를 늘려야 함)
```

# K겹(폴드) 교차 검증

* 데이터가 충분히 많아야 딥러닝 모델의 성능도 향상되는 것이 자명하다.

* 하지만 데이터를 추가적으로 확보하는 것이 쉽지 않을 경우가 있을 것이다.

* 학습 세트를 70%, 테스트 세트를 30%로 분할한 경우,
<br> 30%의 테스트 세트는 학습에 이용할 수 없다는 단점이 있다.

* 이를 해결하기 위해 고안된 방법이 **k겹 교차 검증(k-fold cross validation)**이다.

* k겹 교차 검증이란 데이터셋을 여러 개로 나누어 하나씩 테스트 세트로 사용하고
<br> 나머지를 모두 합해 학습 세트로 사용하는 방법이다.
<br> 이렇게 하면 보유 데이터의 100%를 학습 세트로 사용할 수 있고,
<br> 동시에 테스트 세트로도 사용 가능하다.

* EX. 훈련 세트와 테스트 세트가 각각 4 : 1이라면.
<br> 5개의 폴드가 있다고 했을 때 [4, 1], [3, 1, 1], [2, 1, 2], [1, 1, 3], [1, 4] 총 5개로 나눌 수 있다.

```python
# 5겹 교차 검증 실행(예제)
from sklearn.model_selection import KFold

k = 5
kfold = KFold(n_splits=k, shuffle=True)
acc_score = []

for train_index, test_index in kfold.split(X):
  X_train, X_test = X.iloc[train_index, :], X.iloc[test_index, :]
  y_train, y_test = y.iloc[train_index], y.iloc[test_index]

accuracy = model.evaluate(X_test, y_test)[1]
acc_score.append(accuracy)
```

```python
# 전체 코드
k=5
kfold = KFold(n_splits=k, shuffle=True)
acc_score = []

# 모델 생성
def model_fn():
    model = Sequential()
    model.add(Dense(24, input_dim=60, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    return model

# K겹 교차 검증을 이용해 k번의 학습을 실행
for train_index , test_index in kfold.split(X): # for문으로 k겹 교차 검증의 세트 분할 실행
    X_train , X_test = X.iloc[train_index,:], X.iloc[test_index,:]
    y_train , y_test = y.iloc[train_index], y.iloc[test_index]

    model = model_fn()
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    history = model.fit(X_train, y_train, epochs=200, batch_size=10, verbose=0)
    # verbose=0은 함수 실행시 발생하는 상세 정보들을 출력하지 않겠다는 의미

    accuracy = model.evaluate(X_test, y_test)[1]
    acc_score.append(accuracy)

train_score = history.history['accuracy'][-1]
#k번 실시된 정확도의 평균 계산
avg_acc_score = sum(acc_score)/k

#결과 출력
print('훈련 세트 정확도:', train_score)
print('테스트 세트 정확도:', acc_score)
print('테스트 세트 정확도 평균:', avg_acc_score)

# 평균적으로 약 84%의 정확도가 측정되었음
# 다음 장에서는 학습을 몇 번 반복할지(epochs) 판단하는 방법을 알아보며 모델 성능을 향상시키고자 한다.
```
